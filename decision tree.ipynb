```python
import pandas as pd
dataset=pd.read_csv('Prog 3-data3.csv')
import math
def entropy(p):
    return sum([-i*math.log(i,2) for i in p])
from collections import Counter
def e_list(a_list):
    cnt=Counter(x for x in a_list)
    print('no and yes classes are:',a_list.name,cnt)
    n=len(a_list)*1.0
    p=[x/n for x in cnt.values()]
    return entropy(p)
def info(df,ind,dep,t=0):
    print("info gain calc of",ind)
    df_split=df.groupby(ind)
    for name,group in df_split:
        print(name)
        print(group)
    nobs=len(df.index)*1.0
    df_agg_ent=df_split.agg({dep:[e_list,lambda x:len(x)/nobs]})[dep]
    df_agg_ent.columns=['entropy','propobservation']
    new_entropy=sum(df_agg_ent['entropy']*df_agg_ent['propobservation'])
    old_entropy=e_list(df[dep])
    return old_entropy-new_entropy
print("total entropy of given answer in dataset:",e_list(dataset['Answer']))
print("info gain for outlook is:"+str(info(dataset,'Outlook','Answer')),"\n")
print("info gain for humidity is:"+str(info(dataset,'Humidity','Answer')),"\n")
print("info gain for wind is:"+str(info(dataset,'Wind','Answer')),"\n")
print("info gain for temperature is:"+str(info(dataset,'Temperature','Answer')),"\n")
```
